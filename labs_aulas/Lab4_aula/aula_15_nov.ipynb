```python
import matplotlib.pyplot as plt
import numpy as np
import pickle
from sklearn.metrics import confusion_matrix
import scipy.linalg as la
import sklearn.datasets as ds
```

```python
D=pickle.load(open('RegressData001.p', 'rb'))
(x,y)=(D['x'],D['y'])
# Construir X (2 × 500) - 2ª linha s ´o com “1s”
X=np.vstack((x,np.ones((1,500))))
Rx=np.dot(X,X.T) # matrix 2 × 2
rxy=np.dot(X,y.T) # vector 2 × 1
# Estimar pesos
w=np.dot(np.linalg.pinv(Rx),rxy)
print(w)

plt.figure()
plt.plot(x, y, 'b.')
plt.grid()
plt.show()
```

```python
#Calculo das matrizes de otimização da regressão
n_points = x.shape[1]
X=np.vstack((np.ones((1,n_points)), x))
Rx=np.dot(X,X.T) # matrix 2 × 2
rxy=np.dot(X,y.T) # vector 2 × 1

#Estimar Wopt para a reta de regressão
w = np.dot(np.linalg.pinv(Rx), rxy)
print(w)
```

```python
#Dois pontos para desenhar a reta de regressão
x_min = np.min(x)
x_max = np.max(X)
p1 = np.array([x_min-1, x_max+1])
p1_temp = np.vstack((np.ones([1,2]), p1))
p2 = np.dot(w.T, p1_temp)
P = np.vstack((p1, p2))

plt.figure()
plt.plot(x, y, 'b.')
plt.plot(P[0,:], P[1,:], 'r-')
plt.grid()
plt.show()
```

```python
yh = np.dot(w.T, X)
erros = y-yh
x.shape

plt.figure()
plt.axis((x_min-1., x_max+1., np.min(erros)*3, np.max(erros)*3))
plt.plot(x, erros, 'k.')
plt.grid()
plt.show()
```

```python
#Calculo das matrizes de otimização da regressão
n_points = x.shape[1]
X=np.vstack((np.ones((1,n_points)), x**3, x**2, x))
Rx=np.dot(X,X.T) # matrix 2 × 2
rxy=np.dot(X,y.T) # vector 2 × 1

#Estimar Wopt para a reta de regressão
w = np.dot(np.linalg.pinv(Rx), rxy)
print(w)
```

```python
#Pontos para desenhar a reta de regressão
x_min = np.min(x)
x_max = np.max(x)
p1 = np.arange(x_min-1, x_max+1, 0.25)
n_points = p1.size
p1_temp = np.vstack((np.ones(n_points), p1**3, p1**2, p1))
p2 = np.dot(w.T, p1_temp)
P = np.vstack((p1, p2))

plt.figure()
plt.plot(x, y, 'b.')
plt.plot(P[0,:], P[1,:], 'r-')
plt.grid()
plt.show()
```

```python
yh = np.dot(w.T, X)
erros = y-yh
x.shape

plt.figure()
plt.axis((x_min-1., x_max+1., np.min(erros)*3, np.max(erros)*3))
plt.plot(x, erros, 'k.')
plt.grid()
plt.show()
```
